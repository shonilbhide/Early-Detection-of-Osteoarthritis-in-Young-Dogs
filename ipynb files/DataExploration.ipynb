```python
import os
import pandas as pd

# Define the main directory path
main_directory = "/content/drive/Shareddrives/NNDL_Project"

# Initialize an empty dictionary to store DataFrames
dataframes = {}
columns = ['Frame', 'Time', 'Absolute_Time', 'Row', 'Col', 'Raw_Sum']
# Iterate over directories in the main directory
for root, dirs, files in os.walk(main_directory):
    for dir_name in dirs:
        dataframes[dir_name] = {}
        dir_path = os.path.join(root, dir_name)  # Get the full directory path
        # Iterate over files in the current directory
        for filename in os.listdir(dir_path):
            if filename.endswith("_C.csv"):  # Check if the file name ends with "_C.csv"
                file_path = os.path.join(dir_path, filename)  # Get the full file path
                print(file_path)
                # Read the CSV file into a DataFrame and store it in the dictionary
                dataframes[dir_name][filename]= pd.read_csv(file_path,skiprows=41,names=columns,skipfooter=1)
                #Remove rows with Rawsum equal to zero
                dataframes[dir_name][filename] = dataframes[dir_name][filename][(dataframes[dir_name][filename]['Raw_Sum'] > 0)]

# Print the keys (file names) in the dictionary
print("CSV files ending with '_C' loaded into separate DataFrames:")
print(list(dataframes.keys()))
```

### 3. Comment on the Data:


## Data Imbalance and Characteristics

### Data Distribution:
- The dataset consists of records from 74 normal dogs, 30 dogs with cOA1, and 19 dogs with rOA.
- All dogs belong to age groups ranging from 0 to 5 years, with a significant proportion falling within the 2 to 4 years age range, which correlates with positive osteoarthritis records.

### Pressure Measurements:
- Each dog's data includes variable numbers of runs, with each run capturing pressure measurements from the dog's paws.
- The pressure measurements are captured over a sequence of frames, forming a time-series data. Each frame records the spatial pressure distribution in the form of coordinates on a sensor mat of size 384x88 pixels.

### Data Disparities:
1. **Variable Runs and Frames:**
   - Each dog's data contains a variable number of runs, and within each run, there are variable numbers of frames.
   - Furthermore, the direction of the dog during the runs is inconsistent, leading to variability in the recorded data.

2. **Initial and Final Frames Cleaning:**
   - Runs often include initial and final frames without any pressure readings, which necessitated a cleaning process to remove such instances.

### Lack of Detectible Patterns:
- Despite extensive exploratory data analysis (EDA), the Center of Pressure (CoP) data does not exhibit any discernible patterns, indicating potential challenges in identifying significant features or trends.
